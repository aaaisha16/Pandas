## Exploring olympic speed skating data

```python
import pandas as pd 

df = pd.read_csv("speed_skaters.csv")
df_skaters = df.fillna(0)


df2006 = pd.read_csv("mens1500_2006.csv")
df2010 = pd.read_csv("mens1500_2010.csv")
df2014 = pd.read_csv("mens1500_2014.csv")
df2018 = pd.read_csv("mens1500_2018.csv")
df2022 = pd.read_csv("mens1500_2022.csv")

df_results = pd.concat([df2006, df2010, df2014, df2018, df2022])
df_results
```

```python
df_merge = df_results.merge(df_skaters, left_on = "Athlete", right_on = "Name")
df_merge
```

```python
df_filter = df_merge[(df_merge["Country"] == "USA") & (df_merge["Birth year"] > 1985)]
df_filter
```

We filtered this DataFrame by using the index operator to select all of the athletes that are from the United States. We filtered the DataFrame again by selecting all of the athletes that were born after the year 1985. 

```python
df_weight = df_filter.groupby("Birth year")["Weight"].max()
df_weight
```

We grouped the DataFrame by the athlete's maximum weight for each birth year. For example, in the year 2000, there was no one born so the maximum weight for the athletes was 0.0.
